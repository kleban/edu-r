# Simple Linear Regression

```r
# --- Дані ---
pizza_data <- data.frame(
  Size = c(12, 20, 23, 24, 26, 28, 30, 40),
  Price = c(85, 100, 105, 130, 180, 230, 250, 275)
)
pizza_data
```

```r
# --- Модель лінійної регресії ---
model <- lm(Price ~ Size, data = pizza_data)
```

```r
# --- Результати ---
summary(model)
```

```r
# --- Візуалізація ---
plot(pizza_data$Size, pizza_data$Price,
     main = "Лінійна регресія: Ціна vs Розмір піци",
     xlab = "Розмір піци (см)",
     ylab = "Ціна (грн)",
     pch = 19, col = "blue")
abline(model, col = "red", lwd = 2)
```
