## grouping

```r
## for 2002 avg lifeExp by continent + calculate countries count
```

```r
gapminder |>
  filter(year == 2002) |>
  group_by(continent) |>
  summarise(
    lifeExp2002 = mean(lifeExp),
    countriesCount = n()
  )
```

```r
gapminder |>
  filter(year == 2002) |>
  group_by(continent) |>
  summarise(
    lifeExp2002 = mean(lifeExp),
    countriesCount = n(),
    totalPop = sum(pop)
  )
```

```r
data = gapminder |>
  filter(year == 2002) |>
  group_by(continent) |>
  summarise(
    lifeExp2002 = mean(lifeExp),
    countriesCount = n(),
    totalPop = sum(pop)) |>
    mutate(percPop = round(100*totalPop/sum(gapminder |> filter(year == 2002) |> select(pop) |> sum()), 2))
```

```r
data
```

```r
data |> select(percPop) |> sum()
```

## binding

```r
d2002 <- gapminder |>
            filter(year == 2002) %>% 
            group_by(continent, year) %>% 
            summarise(
                lifeExpAvg = mean(lifeExp),
                countriesCount = n()
            )
head(d2002)
```

```r
d2007 <- gapminder |>
            filter(year == 2007) %>% 
            group_by(continent, year) %>% 
            summarise(
                lifeExpAvg = mean(lifeExp),
                countriesCount = n()
            )
head(d2007)
```

```r
d2002 |> bind_rows(d2007)
```

```r
head(d2002)
```

```r
d2002_new <- gapminder |>
  filter(year == 2002) %>% 
        group_by(continent, year) %>% 
        summarise(
            population = sum(pop),
        )
d2002_new
```

```r
d2002_new <- d2002_new |> select(-c(continent, year))
```

```r
d2002_new
```

```r
d2002 |> bind_cols(d2002_new)
```

```r
d2002_new <- d2002_new |>
  arrange(population)
d2002_new
```

```r
d2002 |> bind_cols(d2002_new |> dplyr::select(-continent, -year)) |>
  select(-continent...5) |>
  rename(continent = continent...1)
```

## join

```r
d2002 <- gapminder %>%
            filter(year == 2002) %>% # year
            group_by(continent, year) %>% # grouping condition
            summarise(
                lifeExpAvg = mean(lifeExp),
                countriesCount = n(), # n() count of rows in group 
                .groups = 'drop'
            )
d2002 |> head()
```

```r
grouped_data2002pop <- gapminder %>%
    filter(year == 2002) %>% 
    group_by(continent, year) %>% 
    summarise(totalPop = sum(pop)) |>
    arrange(totalPop) |> select(-year)

grouped_data2002pop |> head()
```

```r
joined_table <- d2002 |>
  left_join(grouped_data2002pop, by = "continent")
joined_table
```

```r
not_good_data <- data.frame(Name = c("Nick", "Jake", "Anna", "Jane", "Dina"),
                           q1_2021 = c(12442, 22131, 21343, 22111, 14123),
                           q2_2021 = c(13442, 22871, 20343, 22222, 14456),
                           q3_2021 = c(15482, 22031, 22456, 22444, 14533),
                           q4_2021 = c(14511, 20031, 21741, 22333, 14511),
                           q1_2022 = c(12442, 22131, 21343, 22111, 14123),
                           q2_2022 = c(13442, 22871, 20343, 22222, 14456),
                           q3_2022 = c(15482, 22031, 22356, 22444, 14533),
                           q4_2022 = c(14566, 20031, 21741, 22333, 14511))
not_good_data
```
