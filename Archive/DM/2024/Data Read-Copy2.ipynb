## Data from web

```r
# install.packages("rvest")
```

```r
library(rvest)
```

```r
url <- "https://companiesmarketcap.com/"
```

```r
page <- read_html(url)
```

```r
tables <- html_nodes(page, "table")
tables
```

```r
html_table(tables[1])
```

```r
data <- data.frame(html_table(tables[1], header = TRUE))
data
```

```r
data
```

```r
colnames(data) <- c("Rank", "Name", "MarketCap")
```
